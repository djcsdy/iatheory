```python
from cairosvg import svg2png
import os
import subprocess
abc_filename_prefix = f"../src/music/" 

for abc_filename, part in parts:
    abc_file = abc_filename_prefix + abc_filename + '.abc'
    with open(abc_file,'w') as output_file:
        output_file.write(part)
    svg_file = abc_filename_prefix + f"{abc_filename}.svg"
    r = subprocess.run(["abcm2ps", "-g", abc_file, "-O", svg_file])

    svg_file = abc_filename_prefix + f"{abc_filename}001.svg"
    png_file = abc_filename_prefix + f"{abc_filename}.png"
    svg2png(url=svg_file, write_to=png_file)

    #print(abc_file, svg_file,png_file)
```
